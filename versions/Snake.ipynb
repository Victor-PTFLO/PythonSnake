```python
import pygame
from pygame.locals import *
import random

def on_colision(snake_body):
    st_colision = False
    
    for i in range(1, len(snake_body)-1):
        if snake_body[0][0] == snake_body[i][0] and snake_body[0][1] == snake_body[i][1]:
            st_colision = True
            break

    if snake_body[0][0] >= screen_x or snake_body[0][0] <= 0 or snake_body[0][1] >= screen_y or snake_body[0][1] <= 0:
        st_colision = True
    
    return st_colision


def randomize_food():
    x1 = random.randint(0, screen_x) // 10 * 10
    y1 = random.randint(0, screen_y) // 10 * 10

    return x1, y1


# geral game config
screen_x = 500
screen_y = 500
window = pygame.display.set_mode((screen_x, screen_y))
clock = pygame.time.Clock()
default_speed = 5 #speed frame rate
frame_speed = default_speed
comand = pygame.K_LEFT #initial direction

#Snake config
snake_x = screen_x / 2
snake_y = screen_y / 2
default_snake = [(snake_x, snake_y), (snake_x + 10, snake_y), (snake_x + 20,200)]
snake = default_snake.copy()
snake_skin = pygame.Surface((10,10))
snake_skin.fill((0,255,0))

#food position and image
x1, y1 = randomize_food()
food = pygame.Surface((10,10))
food.fill((200,0,100))

#Game run
pygame.init()
open_window = True
while open_window:
    clock.tick(frame_speed)

    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()

        if event.type == KEYDOWN:
            comand = event.key


    if comand == pygame.K_UP:
        snake[0] = (snake[0][0], snake[0][1] - 10)
    if comand == pygame.K_DOWN:
        snake[0] = (snake[0][0], snake[0][1] + 10)
    if comand == pygame.K_LEFT:
        snake[0] = (snake[0][0] - 10, snake[0][1])
    if comand == pygame.K_RIGHT:
        snake[0] = (snake[0][0] + 10, snake[0][1])
    
    if on_colision(snake_body = snake):
        snake = default_snake.copy()
        frame_speed = default_speed
        x1, y1 = randomize_food()
    
    #on eating
    if snake[0][0] == x1 and snake[0][1] == y1:
        x1, y1 = randomize_food()
        snake.append((0,0))
        frame_speed+= 2


    #body movement
    for i in range(len(snake)-1, 0, -1):
        snake[i] = (snake[i-1][0], snake[i-1][1])

    window.fill((0, 0, 0))
    window.blit(food, (x1,y1))
    for pos in snake:
        window.blit(snake_skin, pos)
    
    pygame.display.update()
```
